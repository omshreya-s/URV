```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Copy of user_ank_rain_d1_class_000.txt", sep="\t")
df['Timestamp_sec'] = df['Timestamp [us]'] / 1e6
df.head() 
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['Timestamp_sec'], df['QVAR [LSB]'], color='blue')
plt.xlabel("Time (s)")
plt.ylabel("QVAR [LSB]")
plt.title("QVAR signal over time")
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt
import numpy as np

#define low-pass filter function
def butter_lowpass_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

#parameter setting
fs = 120 
cutoff = 2
order = 4

# take QVAR and time
time_sec = df['Timestamp_sec']
qvar = df['QVAR [LSB]']

qvar_filtered = butter_lowpass_filter(qvar, cutoff, fs, order)
plt.figure(figsize=(12, 6))
plt.plot(time_sec, qvar, label='Raw QVAR', color='blue', alpha=0.5)
plt.plot(time_sec, qvar_filtered, label='Low-pass Filtered QVAR', color='red', linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("QVAR [LSB]")
plt.title(f"QVAR signal - Low-pass filtered (cutoff={cutoff} Hz)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt

def butter_lowpass(data, cutoff, fs, order=4):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low')
    return filtfilt(b, a, data)

fs = 120   # 采样率
cutoff = 2 # 保留 0–2 Hz 生理变化
qvar_lp = butter_lowpass(df['QVAR [LSB]'], cutoff, fs)

def butter_highpass(data, cutoff, fs, order=4):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high')
    return filtfilt(b, a, data)

scr_hp = butter_highpass(qvar_lp, cutoff=0.05, fs=fs)

import matplotlib.pyplot as plt

time = df['Timestamp_sec']

plt.figure(figsize=(12, 5))
plt.plot(time, qvar_lp, label='Tonic (Low-pass)', color='green')
plt.plot(time, scr_hp, label='Phasic (High-pass)', color='orange', alpha=0.7)
plt.xlabel("Time (s)")
plt.ylabel("EDA-like signal")
plt.title("QVAR Signal Processed to EDA-like Components")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, savgol_filter
import pywt

# 假设你已经读取了数据为 df，并含有 QVAR 和时间戳列：
# df['QVAR [LSB]'] 是原始信号
# df['Timestamp_sec'] 是时间轴（单位：秒）

fs = 120  # QVAR 采样率

# 1️⃣ Butterworth 低通滤波
def butter_lowpass(data, cutoff=1.5, fs=120, order=4):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low')
    return filtfilt(b, a, data)

qvar_raw = df['QVAR [LSB]'].values
qvar_lp = butter_lowpass(qvar_raw, cutoff=1.5, fs=fs)

# 2️⃣ Savitzky-Golay 滤波（保留曲线形状）
qvar_sg = savgol_filter(qvar_lp, window_length=31, polyorder=3)

# 3️⃣ 小波去噪
def wavelet_denoise(data, wavelet='db4', level=3, threshold_scale=0.5):
    coeffs = pywt.wavedec(data, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(data))) * threshold_scale
    coeffs[1:] = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs[1:]]
    return pywt.waverec(coeffs, wavelet)

qvar_wdn = wavelet_denoise(qvar_sg)

# ⚠️ 修正小波长度不一致
qvar_wdn = qvar_wdn[:len(qvar_raw)]

# 4️⃣ 指数滑动平均（EWMA）
qvar_ewma = pd.Series(qvar_wdn).ewm(span=25, adjust=False).mean().values

# 5️⃣ 单位归一化到模拟 μS 范围（6.5–9.0）
qvar_min, qvar_max = np.min(qvar_ewma), np.max(qvar_ewma)
qvar_scaled = (qvar_ewma - qvar_min) / (qvar_max - qvar_min)
qvar_us = 6.5 + qvar_scaled * 2.5  # 6.5 – 9.0 µS 范围

# 6️⃣ 画图
plt.figure(figsize=(14, 5))
plt.plot(df['Timestamp_sec'].values, qvar_us, color='green')
plt.title("QVAR Processed to EDA-like Signal (Smoothed, Normalized)")
plt.xlabel("Time (s)")
plt.ylabel("Simulated EDA (µS)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
